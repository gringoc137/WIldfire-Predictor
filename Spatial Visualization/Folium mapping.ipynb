```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt #plotting
import seaborn as sns #for beatiful visualization
import folium
from folium import plugins
```

```python
fire_df = pd.read_csv("C:/Users/Mahe/Desktop/wildfires/final.csv")
fire_df=fire_df.drop(columns=['Unnamed: 0'])
fire_df.head()
```

```python
fire_df.info()
```

```python
topf=fire_df[fire_df['mask']==9.0]
topf
```

```python
m = folium.Map(location=[36.778259,-119.417931], control_scale=True, zoom_start=3,attr = "text some")
df_copy = topf.copy()

# loop through data to create Marker for each hospital
for i in range(0,len(df_copy)):
    
    folium.Marker(
    location=[df_copy.iloc[i]['lat'], df_copy.iloc[i]['lon']],
    #popup=popup,
    tooltip="mask: " + str(df_copy.iloc[i]['mask']) + "<br/> date: "+ str(df_copy.iloc[i]['date']),
    icon=folium.Icon(color='red',icon='fire'),
    ).add_to(m)
        
m
```
