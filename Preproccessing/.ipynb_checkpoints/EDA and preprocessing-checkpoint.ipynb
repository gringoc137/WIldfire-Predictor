```python
plt.scatter(df['lst'],df['mask'], color = 'green') 
  
#plt.plot(X_train, LinReg.predict(X_train), color = 'blue') 
plt.title('LST vs MASK') 
plt.xlabel('lst') 
plt.ylabel('mask') 
  
plt.show()
```

```python
def cal_avg(a):
    avg = a.astype("float").mean(axis=0)
    return avg
```

```python
df['lst'].replace(np.nan, cal_avg(df['lst']), inplace=True)
df['lstn'].replace(np.nan, cal_avg(df['lstn']), inplace=True)
df['ndvi'].replace(np.nan, cal_avg(df['ndvi']), inplace=True)
df['vcr'].replace(np.nan, cal_avg(df['vcr']), inplace=True)
```

```python
df
```

```python
df['vcr'] = df['vcr']/100
```
